```python
pattern = r'(?P<name>[a-zA-Z0-9 \'-]*)( \(((?P<region>.*) - )?.*\))?$'
```

```python
def split(string):
    search = re.search(pattern, string)
    return pd.Series({'name': search.groupdict()['name'], 'region': search.groupdict()['region']})
```

```python
split_dat = dat.SSC_NAME_2016.apply(split)
```

```python
split_dat.head()
```

```python
split_dat.loc[dat.SSC_NAME_2016.str.contains('-')]
```

```python
dat = pd.concat([dat, split_dat], axis=1)
```

```python
dat.head()
```

```python
dat = dat.rename(columns={'STATE_NAME_2016': 'state'})
```
